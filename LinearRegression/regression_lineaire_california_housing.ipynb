```python
import numpy as np
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Chargement du dataset
california_housing=fetch_california_housing(as_frame=True)
```

```python
#Trnasformer le dataset en Frame pour pouvoir le manipuler facilement
import pandas as pd
df = california_housing.frame
```

```python
print(df.head())
```

```python
#Séparation train et test
X=california_housing.data
Y=california_housing.target
feature_names=california_housing.feature_names
class_names=california_housing.target_names
print(feature_names)
print(class_names)
print("Taille de X:",X.shape)
print("Taille de Y:",Y.shape)
(X_train,X_test,Y_train,Y_test)=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#Entainement du modèle
model=LinearRegression()
model.fit(X_train,Y_train)
#Prédiction
Y_pred=model.predict(X_test)
```

```python
#Mesurer la précision des prédictions
mse=mean_squared_error(Y_test,Y_pred)
print(f"Moyenne des erreurs au carré:{mse:.2f}")
rmse=np.sqrt(mse)
print(f"RMSE:{rmse:.2f}")
r2=r2_score(Y_test,Y_pred)
print(f"R2:{r2:.2f}")
```

**Conclusion:**
Le modèle de Régression Linéaire affiche un score $R^2$ de 0,58, ce qui valide sa capacité à capturer la tendance majeure du marché immobilier californien
(le revenu moyen du quartier étant le prédicteur le plus fort).

Toutefois, l'erreur moyenne ($\text{RMSE} = 75\,000\$$) montre les limites d'une approche strictement linéaire.

Pour améliorer ce score, une perspective intéressante serait d'utiliser des algorithmes d'arbres de décision (RandomForestRegressor) capables de prendre en compte les interactions non linéaires, notamment entre la latitude et la longitude (l'effet de l'emplacement géographique sur le prix).
