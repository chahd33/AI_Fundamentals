```python
#Import des bibliohtèques
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
```

```python
#Chargement du dataset
diabete=load_diabetes()
X=diabete.data
Y=diabete.target
print("Taille de X:",X.shape)
print("Taille de Y:",Y.shape)
feature_names=diabete.feature_names
print(feature_names)
#Separation de données train et test
(X_train,X_test,Y_train,Y_test)=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
#Entrainement du modele
model = LinearRegression()
model.fit(X_train,Y_train);
```

```python
#Prédiction
Y_pred=model.predict(X_test)
```

Le **MSE** mesure la moyenne des carrés des erreurs.Pour chaque point,le modèle calcule l'écart entre la réalité et la prédiction,élève cet écart au carré(pour éliminer les signes négatifs et pénaliser plus lourdement les grandes erreurs),puis fait la moyenne de tout cela.

Le **RMSE** est la racine carrée du MSE.

Le **$R^2$ Score** (Coefficient de Détermination) est une métrique normalisée sans unité qui indique à quel point le modèle est plus intelligent qu'une simple ligne droite représentant la moyenne générale de y.

```python
#On veut mesurer la précision des prédictions
#Sachant que la variable cible (Y) est une valeur numérique continue(un sore de progession de la maladie allant de 25 à 346)
mse=mean_squared_error(Y_test,Y_pred)
print(f"MSE:{mse:.2f}(Erreur moyenne au carré)")
rmse=np.sqrt(mse)
print(rmse)
#Le résultat montre que le modèle trompe de 53 points à peu prés sur une échelle qui va de 25 à 346
r2=r2_score(Y_test,Y_pred)
print(f"R2_score:{r2:.2f}")
#On remarque que R2 est <0 donc le modèle est
```

**Conclusion:**
Le modèle obtient un $R^2$ d'environ 0,47 et un RMSE d'environ 53. Cela signifie que le modèle linéaire parvient à expliquer 47 % de la variabilité de la progression de la maladie. En moyenne, les prédictions du modèle s'écartent de 53 points (sur une échelle allant de 25 à 346) par rapport à la réalité médicale des patients.

**Les limites de l'approche linéaire:**
Bien que le modèle capte avec succès la tendance générale (les scores prédits augmentent globalement lorsque la gravité réelle augmente), un $R^2$ en dessous de 0,50 met en évidence deux éléments cruciaux :

**La non-linéarité des données :** La biologie humaine et l'évolution d'une maladie comme le diabète suivent des règles complexes qui ne peuvent pas être parfaitement résumées par une simple ligne droite.

**Le manque d'informations (Bruit) :** 10 variables ne suffisent pas à cerner toute la complexité du métabolisme d'un patient. Des facteurs majeurs absents du dataset (habitudes alimentaires, prédispositions génétiques, activité physique) limitent mathématiquement la précision maximale atteignable.
