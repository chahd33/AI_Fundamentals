```python
#Import des bibliothèques
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
#Chargement du dataset
breast_cancer=load_breast_cancer()
X=breast_cancer.data
Y=breast_cancer.target
print("Taille de X:",X.shape)
print("Taille de Y:",Y.shape)
feature_names=breast_cancer.feature_names
class_names=breast_cancer.target_names
print("Les variables explicatives:",feature_names)
print("Les noms des classes:",class_names)
import pandas as pd
#pour visualiser les premieres lignes on transforme le dataset en DataFrame
df=pd.DataFrame(X,columns=feature_names)
print(df.head())
```

```python
#Separation de données train et test
(X_train,X_test,Y_train,Y_test)=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
# Mise à l'échelle des données
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Entrainer le modèle LogisticRegression avec les données mises à l'échelle
model_scaled = LogisticRegression(max_iter=1000) #j'ai ajouté max_iter=1000 pour que Le modèle prend le temps de converger mathématiquement vers la meilleure solution possible.
model_scaled.fit(X_train_scaled, Y_train);
```

## Pourquoi on a standarisé les données pour LogisticRegression:
la standardisation des données n’est pas une étape optionnelle mais une nécessité pour exploiter pleinement le potentiel de la régression logistique. 
En ramenant toutes les variables à une moyenne de 0 et un écart-type de 1, elle résout trois défis majeurs : 
- Elle accélère drastiquement la convergence de la descente de gradient
- garantit une application juste et efficace de la régularisation ($L_1$/$L_2$)
- Rend les coefficients finaux directement comparables et interprétables.
C'est l'étape fondamentale qui transforme un modèle instable en un classifieur performant, robuste et explicable.

```python
#Prédiction
Y_pred=model_scaled.predict(X_test_scaled)
```

```python
#Evaluation
accuracy=accuracy_score(Y_test,Y_pred)
print("Accuracy du modele Logistic Regression:",accuracy)
print("Matrice de confusion:",confusion_matrix(Y_test,Y_pred))
print("Rapport de classification:")
print(classification_report(Y_test,Y_pred))
```

# Conclusion
Le modèle de Régression Logistique affiche d'excellentes performances avec une exactitude globale (Accuracy) de 98,25 %. L'ajustement des hyperparamètres notamment l'augmentation du nombre maximal d'itérations (max_iter=1000) a permis de stabiliser la convergence du modèle. 

La matrice de confusion et le rapport de classification confirment la robustesse du classifieur, qui obtient des scores de précision et de rappel (recall) équilibrés et supérieurs à 97 % pour les deux classes (0 et 1). 

Le modèle démontre ainsi une excellente capacité de généralisation sans signe de biais envers l'une des classes.
